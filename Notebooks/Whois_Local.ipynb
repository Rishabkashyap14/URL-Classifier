```python
def is_registered(domain_name):
    """A function that returns a boolean indicating whether a `domain_name` is registered"""
    try:
        w = whois.whois(domain_name)
    except Exception:
        return False
    else:
        return w.domain_name, w.creation_date, w.expiration_date
```

```python
whoisLegi=pd.DataFrame(columns=['URL','whois','creation','expiration'])
```

```python
dfLegi=pd.read_csv('DMOZ.csv').sample(2000)
dfLegi
```

```python
count=0
for i in dfLegi.URL:
  domain = urlparse(i).netloc
  if(is_registered(domain)):
    w=is_registered(domain)
    whoisLegi.loc[len(whoisLegi.index)] = [i,w,w[1],w[2]]
    count+=1
    if(count%10==0):
      print(count)
whoisLegi
```
